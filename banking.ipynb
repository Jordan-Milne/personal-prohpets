```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

from fbprophet import Prophet
```

```python
df = pd.read_csv('banking_adj.csv')
```

## Setting the correct format for Prophet

```python
df['cumsum'] = df['amount'].cumsum()
```

```python
purchases = df[['date', 'cumsum']]
purchases.columns = ['ds', 'y']
```

## Fitting Model

```python
m = Prophet(daily_seasonality=False)
m.fit(purchases)
```

## Predicting Future Spending

```python
future = m.make_future_dataframe(periods=365)
```

```python
forecast = m.predict(future)
```

```python
m.plot(forecast);
```

```python
m.plot_components(forecast);
```
